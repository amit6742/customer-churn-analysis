```python
# Sample DataFrame
# df = pd.read_csv('your_dataset.csv')  # Use your own DataFrame
# For demonstration:
data = {'SeniorCitizen': [0, 1, 0, 1, 0, 1, 0, 1],
        'Churn': ['Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No']}
df = pd.DataFrame(data)

# Create a pivot table to calculate the count for each combination
pivot = df.pivot_table(index='SeniorCitizen', columns='Churn', 
                       aggfunc='size', fill_value=0)

# Calculate the percentage for each category
pivot_percentage = pivot.div(pivot.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
pivot_percentage.plot(kind='bar', stacked=True, colormap='viridis')

# Add percentage labels to each bar segment
for i, row in enumerate(pivot_percentage.values):
    for j, value in enumerate(row):
        plt.text(j - 0.15, row[:j].sum() + value / 2, 
                 f'{value:.1f}%', ha='center', va='center', color='white')

# Title and labels
plt.title('Churn by Senior Citizen')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')
plt.legend(title='Churn')
plt.show()
```

```python
# Example data (replace this with your actual DataFrame)
# df = pd.read_csv('your_data.csv')  # Use this if you are reading from a CSV file

# Assuming your DataFrame is already loaded as 'df'
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define the layout for subplots (3x3 grid)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Count Plots for Different Services', fontsize=16)

# Flatten axes for easy iteration
axes = axes.flatten()

# Create a count plot for each column
for idx, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[idx],hue = df["Churn"])
    axes[idx].set_title(f'Count of {col}')
    axes[idx].set_xlabel('')
    axes[idx].set_ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.95])



# Show the plots
plt.show()
```

Most services show a higher count of non-churning customers compared to churning ones. Some features, such as having no internet service, seem associated with lower churn counts, while other services like "OnlineSecurity" and "TechSupport" show a relatively balanced churn pattern.
